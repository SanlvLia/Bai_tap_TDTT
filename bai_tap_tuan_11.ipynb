Bai 3

```python
def sum(n:int):
    s=0
    while(n):
        s+=(n%10)**2
        n//=10
    return s

a=input()
d=set()
kt=0
try:
    x=int(a)
    d.add(x)
    while(x!=1):
        x=sum(x)
        if(x in d):
            kt=1
            print("NO")
            break
        d.add(x)
    if(kt==0):
        print("YES")
except ValueError:
    print("ValueError")
```

```python
a=list(map(int,input().split()))
s=set()
sum=0
for i in range(1,len(a)):
    if(a[i] in s):
        sum=0
        print("Mang khong hop le")
        break
    s.add(a[i])
    sum+=a[i]
if(sum!=0):
    print(sum)
```
