```python

def tk(arr,left:int,right:int,target:int):
    while(left<=right):
        mid=(left+right)//2
        if(arr[mid]==target):
             return mid
        elif(arr[mid]<target):
            left=mid+1
        else:
            right=mid-1
    return -1

res=0

a=list(map(int,input().split()))
x=int(input())
a.sort()
d={}
for i in range(len(a)):
    d[a[i]]=d.get(a[i],0)+1
for i in range(len(a)):
    t=tk(a,0,len(a)-1,x-a[i])
    if(t!=-1):
        res+=d[a[t]]
print(res//2)
```
