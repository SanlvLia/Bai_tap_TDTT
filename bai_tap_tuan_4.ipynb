```python
def kt1(a):
    if("0"<=a<="9"):
        return True
    return False

def kt2(a):
    if("a"<=a<="z"):
        return True
    return False

def kt3(a):
    if("A"<=a<="Z"):
        return True
    return False

s=input()
if(len(s)<=6):
    print("No")
else:
    c1,c2,c3,c4=0,0,0,0
    for i in range(0,len(s)):
        if(kt1(s[i])):
            c1=1
        if(kt2(s[i])):
            c2=1
        if(kt3(s[i])):
            c3=1
        if(kt1(s[i])==False)&(kt2(s[i])==False)&(kt3(s[i])==False):
            c4=1
    if(c1==c2==c3==c4==1):
        print("YES")
    else:
        print("NO")
```
