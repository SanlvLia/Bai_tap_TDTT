```python
a=list(map(int,input().split()))
k=int(input())
res=[]
for i in range(len(a)-1):
    for j in range(i+1,len(a)):
        if(a[j]==k-a[i]):
            res.append((a[i],a[j]))
res.sort()
print(res)
```

```python
a=list(map(int,input().split()))
cnt=[0]*1000000
mx=0
cur=1000000
for i in range(len(a)):
    cnt[a[i]]+=1
    if(cnt[a[i]]>mx):
        mx=cnt[a[i]]
        cur=a[i]
    if(cnt[a[i]]==mx)&(a[i]<cur):
        mx=cnt[a[i]]
        cur=a[i]
print(cur)
```

```python
n,m=map(int,input().split())
k=int(input())
a=[]
s1,s2=0,0
for i in range(n):
    row=list(map(int,input().split()))
    a.append(row)
for i in range(m):
    s1+=a[k-1][i]
for i in range(n):
    s2+=a[i][k-1]
print(s1," ",s2)
```
