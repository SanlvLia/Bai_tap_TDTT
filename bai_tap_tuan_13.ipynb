```python
def reverse(path:str)->int:
    res=0
    with open(path,"r") as f:
        n=f.read()
    for i in range(len(n)-1,-1,-1):
        res=res*10+int(n[i])
    with open(path,"w") as f:
        f.write(str(res))
        f.close()
    return res
s=input()
with open("input.txt","w") as f:
    f.write(s)
    f.close()
reverse("input.txt")
    
```

```python
def areAnagram(path1:str,path2:str):
    with open(path1,"r") as f:
        n1=list(f.read())
    with open(path2,"r") as f:
        n2=list(f.read())
    n1.sort()
    n2.sort()
    if(n1==n2):
        print("True")
    else:
        print("False")
    return
s1=input()
s2=input()
with open("input1.txt","w") as f:
    f.write(s1)
    f.close()
with open("input2.txt","w") as f:
    f.write(s2)
    f.close()
areAnagram("input1.txt","input2.txt")
```

```python
def count(path:str,letter:str):
    cnt=0
    with open(path,"r") as f:
        n=f.read()
    for i in n:
        if(i==letter) or (ord(i)==ord(letter)-32) or (ord(i)==ord(letter)+32):
            cnt+=1
    return cnt
s=input()
k=input()
with open("input.txt","w") as f:
    f.write(s)
print(count("input.txt",k))
```

```python
def fibonacci_to_file(a, b, n, filename="output.txt"):
    if n < 0 or n > 20:
        return
    fib = []
    if n >= 0:
        fib.append(a)
    if n >= 1:
        fib.append(b)
    for i in range(2, n + 1):
        fib.append(fib[i - 1] + fib[i - 2])
    with open(filename, "w", encoding="utf-8") as f:
        f.write(" ".join(map(str, fib)))
```

```python
def maximumProduct(path: str) -> int:
    try:
        with open(path, "r", encoding="utf-8") as f:
            data = list(map(int, f.read().split()))
    except:
        return 0

    n = data[0]
    arr = data[1:]
    arr.sort()
    return max(arr[-1] * arr[-2] * arr[-3],
               arr[0] * arr[1] * arr[-1])
```

```python
def productExceptSelf(path: str) -> list[int]:
    with open(path, "r", encoding="utf-8") as f:
        data = list(map(int, f.read().split()))

    n = data[0]
    nums = data[1:]

    res = [1] * n

    prefix = 1
    for i in range(n):
        res[i] = prefix
        prefix *= nums[i]

    suffix = 1
    for i in range(n - 1, -1, -1):
        res[i] *= suffix
        suffix *= nums[i]

    print(res)
    return res
```

```python
import math

def estimatedTime(path: str) -> float:
    with open(path, "r", encoding="utf-8") as f:
        lines = f.readlines()

    src = des = None
    velocity = 0.0

    for line in lines:
        line = line.strip()
        if line.startswith("SRC"):
            _, coords = line.split(":")
            x, y = map(float, coords.split())
            src = (x, y)
        elif line.startswith("DES"):
            _, coords = line.split(":")
            x, y = map(float, coords.split())
            des = (x, y)
        elif line.startswith("VELOCITY"):
            velocity = float(line.split(":")[1].replace("KM/S", "").strip())

    distance = math.sqrt((des[0] - src[0])**2 + (des[1] - src[1])**2)
    time = distance / velocity

    return round(time, 2)
```

```python
import struct

def read_binary_double():
    path = input().strip()
    with open(path, "rb") as f:
        data = f.read()
        nums = struct.unpack('d' * (len(data) // 8), data)
        for x in nums:
            print(f"{x:.5f}")
```

```python
import math

def imageSmoother(path: str):
    with open(path, "r", encoding="utf-8") as f:
        data = f.read().split()

    idx = 0
    M = int(data[idx]); idx += 1

    img = []
    for _ in range(M):
        row = list(map(int, data[idx:idx + M]))
        idx += M
        img.append(row)

    res = [[0]*M for _ in range(M)]
    dirs = [(-1,-1), (-1,0), (-1,1),
            (0,-1),  (0,0),  (0,1),
            (1,-1),  (1,0),  (1,1)]

    for i in range(M):
        for j in range(M):
            s = cnt = 0
            for dx, dy in dirs:
                ni, nj = i + dx, j + dy
                if 0 <= ni < M and 0 <= nj < M:
                    s += img[ni][nj]
                    cnt += 1
            res[i][j] = s // cnt

    for row in res:
        print(" ".join(map(str, row)))
```
