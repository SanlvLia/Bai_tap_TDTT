```python
class cylinder:
    def __init__(self,r,h):
        self.r=r
        self.h=h
    def S_be_mat(self):
        return 2 * 3.14 * self.r * (self.r + self.h)
    def V(self):
        return 3.14 * self.r * self.r * self.h
    
r,h=map(float,input().split())
c=cylinder(r,h)
print(round(c.S_be_mat(),2),round(c.V(),2))
```

```python
class date:
    def __init__(self,d,m,y):
        self.d=d
        self.m=m
        self.y=y
    def cal(self):
        if(self.d==31):
            if(self.m==12):
                return([1,1,self.y+1])
            else:
                return([1,self.m+1,self.y])
        if(self.d==30)and(self.m in [4,6,9,11]):
            return([1,self.m+1,self.y])
        if(self.m==2):
            if((self.y%400!=0)and(self.d==28))or((self.y%400==0)and(self.d==29)):
                return(1,3,self.y)
            else:
                return([self.d+1,2,self.y])
        return([self.d+1,self.m,self.y])
    def kt(self):
        if(self.y<0):
            return("INVALID")
        if(self.m<0)or(self.m>12):
            return("INVALID")
        if(self.m in [2,4,6,9,11])and(self.d>30):
            return("INVALID")
        if(self.m==2)and(self.y%400!=0):
            if(self.d>28):
                return("INVALID")
dd,mm,yyyy=input().split("/")
dd,mm,yyyy=int(dd),int(mm),int(yyyy)
x=date(dd,mm,yyyy)
if(x.kt()=="INVALID"):
    print(x.kt())
else:
    a=x.cal()
    for i in range(len(a)-1):
        print(a[i],end="/")
    print(a[len(a)-1])
        
```

```python
import math
class point:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def kt(self):
        if(self.x==0)and(self.y==0):
            return("Gốc tọa độ")
        if(self.x==0):
            return("Trục tung")
        if(self.y==0):
            return("Trục hoành")
    def cal(self):
        res=math.sqrt(self.x*self.x+self.y*self.y)
        return(round(res,2))
x,y=map(float,input().split())
p=point(x,y)
print(p.kt())
print(p.cal())
```

```python
class caculator:
    def __init__(self,a,x,b):
        self.a=a
        self.x=x
        self.b=b
    def cal(self):
        if(self.x=="+"):
            return self.a+self.b
        if(self.x=="-"):
            return self.a-self.b
        if(self.x=="x"):
            return self.a*self.b
        if(self.x=="/"):
            return self.a/self.b
        if(self.x=="^"):
            return self.a**self.b
        if(self.x=="%"):
            return self.a%self.b
while(True):
    print("Bạn có muốn thoát ?")
    s=list(input().split())
    if(s[0]=="YES")or(s[0]=="yes")or(s[0]=="Yes"):
        break
    try:
        s[0],s[2]=int(s[0]),int(s[2])
        x=caculator(s[0],s[1],s[2])
        print(x.cal())
    except:
        continue



```

```python
class rectangle:
    def __init__(self,w,h):
        self.w=w
        self.h=h
    def area(self):
        return self.w*self.h
    def perimeter(self):
        return 2*(self.w+self.h)
    def scale(self,k):
        self.k=k
        self.w*=self.k
        self.h*=self.k
w,h,k=map(float,input().split())
x=rectangle(w,h)
x.scale(k)
print(round(x.area(),2),round(x.perimeter(),2))
```

```python
from math import gcd
class fraction:
    def __init__(self,num,den):
        if (den==0):
            raise ValueError("")
        if(den<0):
            num=-num
            den=-den
        g=gcd(abs(num),den)
        self.num=num//g
        self.den=den//g
    def add(self,other):
        return fraction(
        self.num * other.den + other.num * self.den
    ) 
    
```
